```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy import interpolate
%matplotlib inline
```

```python
z=np.array([[5,4,3],[1,2,4]])
print(z.shape)
x=[1,2]
y=[4,5,6]

f = interpolate.interp2d(x, y, z.T)
xi=np.arange(0,3.5,0.5)
yi=np.arange(3,7,0.5)
azi=f(xi,yi)
plt.contourf(x,y,z.T)
plt.colorbar()
plt.xlim([0,3])
plt.show()
plt.contourf(xi,yi,azi)
plt.colorbar()
```
